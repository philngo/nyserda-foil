Analysis of NYSERDA Energy efficiency programs
==============================================

Summary statistics, charts, etc.

```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.dates import date2num,DateFormatter
```

Load the data

```python
data = pd.read_excel("FOIL 8.2013.xlsx")
```

Start exploring

```python
data.count()
```

```python
data.mean() # some of these are meaningless, like zipcode, project id. But the rest of them work.
```

Interesting means:

- SIR: 1.51
- Total costs: \$8000
- Financed amount: \$1660

It would be nice to look at some of these in a bit more detail. What are the distributions?

```python
plt.hist(data["SQFT"],bins=100)
plt.title("Square footage (program-wide)")
plt.show()
```

```python
plt.hist(date2num(data["COMPLETED"]),bins=70)
plt.title("Completion dates (program-wide)")
ax=plt.gca()
xfmt = DateFormatter('%Y-%m')
ax.xaxis.set_major_formatter(xfmt)
plt.xticks( rotation=25 )
plt.show()
```

```python
plt.hist(data["TOTAL COSTS"],bins=100)
plt.title("Project costs (program-wide)")
plt.show()
```

```python
companies = {}
for name,group in data.groupby("COMPANY NAME"):
    companies[name] = group
```

```python
plt.hist([ companies[name]["Project ID"].count() for name in companies.keys()],bins=70)
plt.title("Projects per company")
plt.show()
```

```python
plt.hist(data["PRE CFM50"].dropna().values,bins=np.arange(0, 15000, 500),alpha=.6)
plt.hist(data["PROP CFM50"].dropna().values,bins=np.arange(0, 15000, 500),alpha=.6)
plt.hist(data["POST CFM50"].dropna().values,bins=np.arange(0, 15000, 500),alpha=.6)
plt.title("Blower Door pre/prop/post (program-wide)")
plt.legend(["Pre","Prop","Post"])
plt.show()
```

```python
plt.hist(data["PRE KWH USAGE"],bins=np.arange(0, 40000, 500),alpha=.6)
plt.hist(data["POST KWH USAGE"],bins=np.arange(0, 40000, 500),alpha=.6)
plt.title("kWh usage pre/post (program-wide)")
plt.legend(["Pre","Post"])
plt.show()
```

```python
plt.hist(data["PRE MMBTU USAGE"],bins=np.arange(0, 1000, 10),alpha=.6)
plt.hist(data["POST MMBTU USAGE"],bins=np.arange(0, 1000, 10),alpha=.6)
plt.title("MMBTU usage pre/post (program-wide)")
plt.legend(["Pre","Post"])
plt.show()
```
