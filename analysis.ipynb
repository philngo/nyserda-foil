Analysis of NYSERDA FOIL
========================



```python
%matplotlib inline

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("FOIL 8.2013.xlsx")
```

```python
data.count()
```

```python
data["TOTAL COSTS"].mean()
```

```python
data["FINANCED AMT"].mean()
```

```python
plt.scatter(data["TOTAL COSTS"],data["FINANCED AMT"], edgecolors='none',alpha=.01)
plt.xlabel("Total Costs")
plt.ylabel("Financed amount")
plt.title("Project financing tracks")
plt.show()
```

```python
plt.scatter(data["TOTAL COSTS"],data["TOTAL GRANT"], edgecolors='none',alpha=.01)
plt.xlabel("Total Costs")
plt.ylabel("Total Grant")
plt.title("Grants recieved")
plt.show()
```

```python
plt.hist(data["TOTAL COSTS"],bins=100)
plt.title("Project costs")
plt.show()
```

```python
plt.hist(data["TOTAL GRANT"],bins=100)
plt.title("Project grants")
plt.show()
```

```python
plt.hist(data["FINANCED AMT"],bins=100)
plt.title("Project financed amounts")
plt.show()
```

```python
plt.hist(data["CASH BACK"],bins=100)
plt.title("Project cash back")
plt.show()
```

```python
plt.hist(data["FY KWH SAVED"],bins=100)
plt.title("Projected first year kWh savings")
plt.show()
```

```python
plt.hist(data["FY MMBTU SAVED"],bins=100)
plt.title("Projected first year mmbtu savings")
plt.show()
```

```python
plt.hist(data["SIR"].dropna().values,bins=100)
plt.title("Software savings to investment ratio")
plt.show()
```

```python
plt.scatter(data["TOTAL COSTS"],data["SIR"],edgecolors="none",alpha=0.02)
plt.title("costs by savings to investment ratio")
plt.show()
```

```python
plt.scatter(data["TOTAL COSTS"],data["FY MMBTU SAVED"],edgecolors="none",alpha=0.01)
plt.title("")
plt.show()
```

```python
plt.scatter(data["PRE KWH USAGE"],data["POST KWH USAGE"],edgecolors="none",alpha=0.01)
plt.title("Modeled KWH usage before and after")
plt.show()
```

```python
plt.scatter(data["PRE MMBTU USAGE"],data["POST MMBTU USAGE"],edgecolors="none",alpha=0.1)
plt.title("Modeled MMBTU usage before and after")
plt.show()
```

```python
plt.scatter(data["PRE MMBTU USAGE"],data["CUST RPT MMBTU"],edgecolors="none",alpha=0.1)
plt.title("Cust report(pre) compared with model(pre)")
plt.show()
```

```python
plt.scatter(data["POST MMBTU USAGE"],data["CUST RPT MMBTU"],edgecolors="none",alpha=0.1)
plt.title("Cust report (pre) compared with model (post)")
plt.show()
```

```python
plt.scatter(data["PRE KWH USAGE"],data["CUST RPT KWH"],edgecolors="none",alpha=0.01)
plt.title("Cust report(pre) compared with model(pre)")
plt.show()
```

```python
plt.scatter(data["POST KWH USAGE"],data["CUST RPT KWH"],edgecolors="none",alpha=0.01)
plt.title("Cust report(pre) compared with model(post)")
plt.show()
```
