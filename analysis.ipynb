Analysis of NYSERDA Energy efficiency programs
==============================================

Summary statistics, charts, etc.

```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
```

Load the data

```python
data = pd.read_excel("FOIL 8.2013.xlsx")
```

Start exploring

```python
data.count()
```

```python
data.mean() # some of these are meaningless, like zipcode, project id, etc. But most of them aren't.
```

Interesting means:

- SIR: 1.51
- Total costs: $8000
- Financed amount: $1660

Interesting, but it would be nice to look at some of these 

```python
plt.scatter(data["TOTAL COSTS"],data["FINANCED AMT"], edgecolors='none',alpha=.01)
plt.xlabel("Total Costs")
plt.ylabel("Financed amount")
plt.title("Project financing tracks (program wide)")
plt.show()
```

```python
plt.scatter(data["TOTAL COSTS"],data["TOTAL GRANT"], edgecolors='none',alpha=.01)
plt.xlabel("Total Costs")
plt.ylabel("Total Grant")
plt.title("Grants recieved (program wide)")
plt.show()
```

```python
plt.hist(data["TOTAL COSTS"],bins=100)
plt.title("Project costs")
plt.show()
```

```python
plt.hist(data["TOTAL GRANT"],bins=100)
plt.title("Project grants")
plt.show()
```

```python
plt.hist(data["FINANCED AMT"],bins=100)
plt.title("Project financed amounts")
plt.show()
```

```python
plt.hist(data["CASH BACK"],bins=100)
plt.title("Project cash back")
plt.show()
```

```python
plt.hist(data["FY KWH SAVED"],bins=100)
plt.title("Projected first year kWh savings")
plt.show()
```

```python
plt.hist(data["FY MMBTU SAVED"],bins=100)
plt.title("Projected first year mmbtu savings")
plt.show()
```

```python
plt.hist(data["SIR"].dropna().values,bins=100)
plt.title("Software savings to investment ratio")
plt.show()
```
